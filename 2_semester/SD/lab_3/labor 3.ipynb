```python
x = var('x')
y = function('y')(x)
eqd = diff(y,x,2)+diff(y,x)==sin(x)+cos(x)
sol = desolve(eqd, y)
sol
```

```python
_K1,_K2=var('_K1,_K2')
sol.substitute(_K1==1,_K2==2)
```

```python
sol1 = sol.substitute(_K1==1,_K2==2)
plot(sol1,x,-2,2)
```

```python
eqd=diff(y,x,2)-diff(y,x)==exp(2*x)
sol=desolve(eqd,y)
sol
```

```python
_K1,_K2=var('_K1,_K2')
sol2 = sol.substitute(_K1==1,_K2==5)
plot(sol2,x,-2,2)
```
